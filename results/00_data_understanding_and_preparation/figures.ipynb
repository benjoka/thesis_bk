```python
import pandas as pd
import plotly.express as px
```

```python
distance_distribution = [
    {"geographic_area": "Düsseldorf", "seasonal_period": "Week", "median_distance": 13, "mean_edge_weight":1.72},    
    {"geographic_area": "Düsseldorf", "seasonal_period": "Month", "median_distance": 9, "mean_edge_weight":2.25},
    {"geographic_area": "Düsseldorf", "seasonal_period": "Year", "median_distance": 25, "mean_edge_weight":4.26},
    {"geographic_area": "NRW", "seasonal_period": "Week", "median_distance": 15, "mean_edge_weight":2.08},    
    {"geographic_area": "NRW", "seasonal_period": "Month", "median_distance": 11, "mean_edge_weight":2.8},
    {"geographic_area": "NRW", "seasonal_period": "Year", "median_distance": 27, "mean_edge_weight":4.85},
    {"geographic_area": "Germany", "seasonal_period": "Week", "median_distance": 15, "mean_edge_weight":2.55},    
    {"geographic_area": "Germany", "seasonal_period": "Month", "median_distance": 10, "mean_edge_weight":2.54},
    {"geographic_area": "Germany", "seasonal_period": "Year", "median_distance": 33, "mean_edge_weight":4.24},
]
```

```python
distance_distribution_df = pd.DataFrame(distance_distribution)
fig = px.bar(distance_distribution_df, x='geographic_area', y="median_distance", color='seasonal_period', barmode='group', color_discrete_sequence=["lightgrey", "darkgrey", "grey"],
             title='Grouped Bar Chart', labels={"seasonal_period": ""})
fig.update_yaxes(type="log")
fig.update_layout(template="presentation", bargap=0.1, bargroupgap=0.1, width=700, height=500,
                  font=dict(size=22),
                  margin=dict(l=120,r=0,t=0,b=40),
                  xaxis=dict(
                        range=[-0.7, 2.7],
                        title=dict(
                            text='',
                            standoff=0
                        ),
                        tickangle=0
                    ),
                  title="",
                    yaxis=dict(
                        title=dict(
                            text='Median distance',
                            standoff=20
                        ),
                        tickangle=0
                    )
                 )
fig.show()
fig.write_image("figures/sequence_counts.svg", width=700, height=500)
```
