```python
infection_distribution_df = pd.DataFrame(infection_distribution)
fig = px.bar(infection_distribution_df, x='geographic_area', y="sequence_count", color='seasonal_period', barmode='group', color_discrete_sequence=["lightgrey", "darkgrey", "grey"],
             title='Grouped Bar Chart', labels={"seasonal_period": ""})
fig.update_yaxes(type="log")
fig.update_layout(template="presentation", bargap=0.1, bargroupgap=0.1, width=700, height=500,
                  font=dict(size=22),
                  margin=dict(l=120,r=0,t=0,b=40),
                  xaxis=dict(
                        range=[-0.7, 2.7],
                        title=dict(
                            text='',
                            standoff=0
                        ),
                        tickangle=0
                    ),
                  title="",
                    yaxis=dict(
                        title=dict(
                            text='Sequence count',
                            standoff=20
                        ),
                        tickangle=0
                    )
                 )
fig.show()
fig.write_image("figures/sequence_counts.svg", width=700, height=500)
```
